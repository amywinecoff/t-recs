```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

n = 1125

# Arrays
a = np.logspace(n/10, 0, num=n, base=2)
# a normalized
prob = a/a.sum()
```

```python
prob
```

```python
a
```

```python
plt.yscale('linear')
plt.scatter(np.arange(n), prob)
```

```python
plt.yscale('linear')
plt.scatter(np.arange(n), a)
```

```python
plt.yscale('log')
plt.scatter(np.arange(n), a)
plt.yscale('log')
plt.scatter(np.arange(n), prob)
```
